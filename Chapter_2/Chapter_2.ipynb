```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist
```

```python
# 一行是一个数据点
X_data = np.array([[3, 3], [4, 3], [1, 1]])
y_label = np.array([+1, +1, -1])
```

```python
X_data, y_label
```

## 感知机随机梯度下降模拟

```python
def plot_bound(w, b, X, y):
    """ 画出 y = w * x + y 的边界及 数据X的点 """
    x_point = np.linspace(-1,5,1000)
    y_point = - (b + w[0] * x_point) / w[1]
    fig = plt.figure(figsize=(8, 8))
    #使用axisartist.Subplot方法创建一个绘图区对象ax
    ax = axisartist.Subplot(fig, 111)  
    #将绘图区对象添加到画布中
    fig.add_axes(ax)

    #通过set_visible方法设置绘图区所有坐标轴隐藏
    ax.axis[:].set_visible(False)

    #ax.new_floating_axis代表添加新的坐标轴
    ax.axis["x"] = ax.new_floating_axis(0,0)
    #给x坐标轴加上箭头
    ax.axis["x"].set_axisline_style("->", size = 1.0)
    #添加y坐标轴，且加上箭头
    ax.axis["y"] = ax.new_floating_axis(1,0)
    ax.axis["y"].set_axisline_style("-|>", size = 1.0)
    #设置x、y轴上刻度显示方向
    ax.axis["x"].set_axis_direction("top")
    ax.axis["y"].set_axis_direction("right")

    #设置x、y坐标轴的范围
    plt.xlim(-1, 5)
    plt.ylim(-1, 5)
    #绘制图形
    for i in range(len(y)):
        c = 'b' if y[i] < 0 else 'r'
        plt.plot(X[i,0], X[i,1], c+'o')
        ax.text(X[i,0],X[i,1]+0.2,'x'+str(i+1),fontsize=12,horizontalalignment='center')
    plt.plot(x_point, y_point, label='board')
```

```python
def f(w:np.array, b:float, X:np.array, y:np.array) -> np.array:
    """ 计算每个 y[i] * (w * x[i] + b) 的值
    :param w: 感知机的权值
    :param b: 感知机的偏置
    :param X: 数据的坐标(每一行一个数据)
    :param y: 数据的标签
    """
    return y * (w.dot(X.T) + b)
```

```python
def sgd(w_init:np.array, b_init:float, X:np.array, y:np.array, lr:float = 1.):
    """使用随机梯度下降法更新 w和b
    :param w_init: w的初始值
    :param b_init: b的初始值
    :param lr: 学习率/步长
    """
    w = w_init
    b = b_init
    finish = False
    epoch = 0
    print('Init[{}]: w = {}, b = {}\n'.format(epoch, w, b))
    while not finish:
        finish = True
        epoch += 1
        L = f(w, b, X, y)
        for i, l in enumerate(L):
            if l <= 0:
                w += lr * y[i] * X[i,:]
                b += lr * y[i]
                print('Epoch[{}]: Choose x_{}, w = {}, b = {}'.format(epoch, i+1, w, b))
                finish = False
                break
    return w, b
```

```python
w_init = np.zeros(X_data.shape[1])
b_init = 0.
w, b = sgd(w_init, b_init, X_data, y_label)

plot_bound(w, b, X_data, y_label)
```

## 感知机的对偶形式随机梯度下降模拟

```python
def get_gram(X:np.array):
    """ 求出Gram矩阵 """
    return X.dot(X.T)
```

```python
def dual_sgd(alpha_init:np.array, b_init:float, X:np.array, y:np.array, lr:float = 1.):
    """ 感知机的对偶形式，使用梯度下降算法更新参数 
    :param alpha_init: alpha_init[i]表示x[i]对应分量
    """
    alpha = alpha_init
    b = b_init
    G = get_gram(X)
    finish = False
    epoch = 0
    print('Init[{}]: alpha = {}, b = {}\n'.format(epoch, alpha, b))
    while not finish:
        finish = True
        epoch += 1
        for i in range(len(y)):
            if y[i] * (np.sum(alpha * y * G[i,:]) + b) <= 0:
                alpha[i] += lr
                b += y[i]
                print('Epoch[{}]: Choose x_{}, alpha = {}, b = {}'.format(epoch, i+1, alpha, b))
                finish = False
                break
    return alpha, b
```

```python
alpha_init = np.zeros(X_data.shape[0])
b_init = 0.
alpha, b = dual_sgd(alpha_init, b_init, X_data, y_label, lr=1.)

w = np.sum(X_data.T * y_label * alpha, axis=1)
plot_bound(w, b, X_data, y_label)
```

```python
# 使用更小的学习率
alpha_init = np.zeros(X_data.shape[0])
b_init = 0.
alpha, b = dual_sgd(alpha_init, b_init, X_data, y_label, lr=0.5)

w = np.sum(X_data.T * y_label * alpha, axis=1)
plot_bound(w, b, X_data, y_label)
```

```python
# 使用不同的初始值
alpha_init = np.ones(X_data.shape[0])
b_init = 0.5
alpha, b = dual_sgd(alpha_init, b_init, X_data, y_label, lr=1.)

w = np.sum(X_data.T * y_label * alpha, axis=1)
plot_bound(w, b, X_data, y_label)
```

可以看到，使用不同的初始值或学习率都可能产生不一样的结果！
